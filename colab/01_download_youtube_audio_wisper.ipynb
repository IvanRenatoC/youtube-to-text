###2.2 Transcribe (y guarda archivo)

```python
import whisper

model = whisper.load_model("small")  # "tiny" mÃ¡s rÃ¡pido, "medium" mejor calidad (mÃ¡s lento)
result = model.transcribe("audio.mp3", language="es")  # si es espaÃ±ol

texto = result["text"]

with open("transcripcion.txt", "w", encoding="utf-8") as f:
    f.write(texto)

texto[:500]
```

##Parte 3 â€” Guardar tambiÃ©n con timestamps (opcional, Ãºtil)

Si quieres texto por segmentos con tiempo:

```python
segments = result["segments"]

with open("transcripcion_con_tiempos.txt", "w", encoding="utf-8") as f:
    for s in segments:
        f.write(f"[{s['start']:.2f} - {s['end']:.2f}] {s['text'].strip()}\n")

segments[:2]
```
